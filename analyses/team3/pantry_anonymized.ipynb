```python
########################
#  Data Normalization  #
#  And Exploration     #
########################

pantryData = pd.read_csv("pantry_anonymized.csv")
pantryData['assistance_date'] = pd.to_datetime(pantryData['assistance_date']) #convert month to DateTime type
pantryData['assistance_category'] = pantryData['assistance_category'].replace(to_replace='Food Pantry: ', value='',regex=True)
print(pantryData['assistance_category'].unique())
#pantryData['assistance_date'].max() #look at latest date in the dataset
pantryData['unit'].value_counts() #look at count of units column values
```

```python
#filter to include pounds only
pantryData = pantryData[pantryData['unit'] == 'Pounds'] #only 100 records with dollars unit
pantryData['assistance_category'].value_counts()
```

```python
#################
# Data Analysis #
#################

listOfCategories = ['Food Pantry Poundage', 'Holiday Baskets']

generateGraphs.generateGraphSet(dataFrameNameForTitle ='Pantry',df= pantryData, dateColumn= 'assistance_date', specialColumnForCategory= 'assistance_category', 
                 listOfCategories = listOfCategories, columnForAnalysis = 'amount', 
                 yAxisNameForGraph = 'Count of Pounds Distributed', isSummed = 'Yes')
```
